```python
from collections import deque
queue = deque(["Ram", "Tarun", "Asif", "John"])
print(queue)   # Mencetak output dari queue
queue.append("Akbar")   # Menambahkan elemen ("Akbar") pada bagian ujung kanan
print(queue)   # Mencetak output dari queue
queue.append("Birbal")   # Menambahkan elemen ("Birbal") pada bagian ujung kanan
print(queue)   # Mencetak output dari queue
print(queue.popleft())   # Menghapus element dari bagian ujung kiri
print(queue.popleft())   # Menghapus elemen dari ujung kiri
print(queue)   # Mencetak output dari queue
```

```python
def TowerOfHanoi(n, dr, menuju, bantuan):
  if n == 0:
    return
  TowerOfHanoi (n-1, dr, bantuan, menuju)
  print ("Pindah Disk", n, "dari", dr, "ke", menuju)
  TowerOfHanoi (n-1, bantuan, dr, menuju)

# Driver Code
N = 3

# A, C, B are the name of rods
TowerOfHanoi(N, 'A', 'C', 'B')
```

```python
def sortStack(input):
  tmpStack = []
  while(len(input) > 0):
    # Mengeluarkan elemen pertama
    tmp = input[-1]
    input.pop

    # Ketika tumpukan sementara tidak kosong
    # Bagian atas tumpukan lebih kecil dari template
    while(len(tmpStack) > 0 and tmpStack[-1] < tmp):

      # Pop dari tumpukan sementara dan tambahkan ke tumpukan input
      input.append(tmpStack[-1])
      tmpStack.pop()

    # Tambahkan template sementara di tumpukan
    tmpStack.append(tmp)

  return tmpStack
```

```python
def sortArrayUsingStacks(arr, n):

  # Tambahkan elemen array ke tumpukan
  input = []
  i = 0
  while (i < n):
    input.append(arr[i])
    i = i + 1

  # Mengurutkan tumpukan sementara
  tmpStack = sortStack(input)
  i = 0

  # Letakkan elemen tumpukan di arrp[]
  while (i < n):
    arr[i] = tmpStack[-1]
    tmpStack.pop()
    i = i + 1

  return arr

# Driver Code
arr = [10, 5, 15, 45]
n = len(arr)

arr = sortArrayUsingStacks(arr, n)
i = 0

while(i < n):
  print(arr[i], end = " ")
  i = i + 1
```
