# **Bubble Sort**

```python
def Bubble_Sort(b):   # Mendefinisikan fungsi Bubble_Sort yang akan menerima angka (arr) sebagai argumen dan akan mengurutkan angka menggunakan Bubble_Sort.
    for i in range(len(b) - 1, 0, -1):
        for j in range(i):
            if b[j] > b[j + 1]:
                temp = b[j]
                b[j] = b[j + 1]
                b[j + 1] = temp

angka = [64, 25, 12, 22, 11]   # Membuat angka dengan nilai yang belum diurutkan.
Bubble_Sort(angka)   # Memanggil fungsi Bubble_Sort dengan angka sebagai argumen untuk mengurutkan angka.
print("Angka yang diurutkan : ")   # Menampilkan pesan untuk menandakan bahwa hasil output berikutnya adalah angka yang sudah diurutkan.
print(angka)   # Menampilkan angka yang telah diurutkan setelah memanggil fungsi Bubble_Sort.
```

# **Selection Sort**

```python
def Selection_Sort(arr):   # Mendefinisikan fungsi Selection_Sort yang akan menerima angka (arr) sebagai argumen dan akan mengurutkan angka menggunakan Selection_Sort.
    n = len(arr)   # Menghitung panjang angka dan menyimpannya dalam variabel n.
    for i in range(n - 1):   # Loop luar akan mengiterasi melalui seluruh elemen angka, kecuali elemen terakhir (n-1).
        min_index = i   # Menetapkan indeks pertama (i) sebagai indeks elemen terkecil dalam angka yang belum diurutkan.
        for j in range(i + 1, n):   # Loop dalam akan mengiterasi melalui sisa angka yang belum diurutkan, dimulai dari indeks setelah i (i+1) hingga indeks terakhir (n).
            if arr[j] < arr[min_index]:   # Memeriksa apakah nilai elemen ke-j lebih kecil daripada nilai elemen terkecil yang ditemukan sejauh ini.
                min_index = j
        arr[i], arr[min_index] = arr[min_index], arr[i]   #  Setelah mencari nilai terkecil dalam sisa angka yang belum diurutkan, nilai terkecil tersebut akan ditukar dengan elemen pada indeks ke-i.

arr = [64, 25, 12, 22, 11]   # Membuat angka dengan nilai yang belum diurutkan.
Selection_Sort(arr)   # Memanggil fungsi Selection_Sort dengan angka sebagai argumen untuk mengurutkan angka.
print("Angka yang diurutkan : ")   # Menampilkan pesan untuk menandakan bahwa hasil output berikutnya adalah angka yang sudah diurutkan.
print(arr)   # Menampilkan angka yang telah diurutkan setelah memanggil fungsi Selection_Sort.
```

# **Insertion Sort**

```python
def InsertionSort(arr):   # Mendefinisikan fungsi Insertion_Sort yang akan menerima angka (arr) sebagai argumen dan akan mengurutkan angka menggunakan Insertiob_Sort.
	n = len(arr)   # Menghitung panjang angka dan menyimpannya dalam variabel n.

	if n <= 1:
		return   # Jika array memiliki 0 atau 1 elemen, maka array sudah diurutkan, jadi kembalikan.

	for i in range(1, n):   # Mengulangi array dari elemen kedua.
		key = arr[i]   # Menyimpan elemen saat ini sebagai kunci untuk dimasukkan pada posisi yang tepat.
		j = i-1
		while j >= 0 and key < arr[j]:   # Pindahkan elemen yang lebih besar dari kunci satu posisi ke depan.
			arr[j+1] = arr[j]   # Geser elemen ke kanan.
			j -= 1
		arr[j+1] = key   # Masukan kunci pada posisi yang benar.

arr = [64, 25, 12, 22, 11]   # Membuat angka dengan nilai yang belum diurutkan.
InsertionSort(arr)   # Memanggil fungsi Insertion_Sort dengan angka sebagai argumen untuk mengurutkan angka.
print("Angka yang diurutkan : ")   # Menampilkan pesan untuk menandakan bahwa hasil output berikutnya adalah angka yang sudah diurutkan.
print(arr)   # Menampilkan angka yang telah diurutkan setelah memanggil fungsi Insertion_Sort.
```
