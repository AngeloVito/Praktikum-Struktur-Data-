# Tuple

```python
tuple_jenis_kelamin = ('Laki-Laki', 'Perempuan')

print(tuple_jenis_kelamin[0]) # indeks satu
print(tuple_jenis_kelamin[-1]) # indeks nol
```

```python
tuple_buah = ('Pisang', 'Nanas', 'Melon', 'Durian')

print(tuple_buah[0:])
print(tuple_buah[1:])
print(tuple_buah[2:])
print(tuple_buah[3:])
print(tuple_buah[:0])
print(tuple_buah[:1])
print(tuple_buah[:2])
print(tuple_buah[:3])
print(tuple_buah[:4])
```

# Dictionary

```python
harga = dict(pensil = 3000, pulpen = 5000) # dictionary
print(harga) # mencetak dictionary dari pensil = 3000, pulpen = 5000
```

```python
harga = {'pensil': 3000, 'pulpen': 5000} # dictionary
print(harga['pensil']) # hanya mencetak harga dari item yang ada di dictionary
print(harga['pulpen'])
```
